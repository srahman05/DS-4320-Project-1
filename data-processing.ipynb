# Processing the Data

The EPA data files (AQI, NO2, and Ozone) are separated by year, so I will merge them and save each "table" as parquet. The fourth "table" comes from the EPA National Emissions Inventory in 2020 and the files are separated by U.S. region. I will merge them and save as parquet. 

## Table 1: Daily AQI by County from 2020 to 2025

```python
import pandas as pd

# read all years
aqi20 = pd.read_csv('daily_aqi_by_county_2020.csv')
aqi21 = pd.read_csv('daily_aqi_by_county_2021.csv')
aqi22 = pd.read_csv('daily_aqi_by_county_2022.csv')
aqi23 = pd.read_csv('daily_aqi_by_county_2023.csv')
aqi24 = pd.read_csv('daily_aqi_by_county_2024.csv')
aqi25 = pd.read_csv('daily_aqi_by_county_2025.csv')

# merge all years
aqi = pd.concat([aqi20, aqi21, aqi22, aqi23, aqi24, aqi25], ignore_index=True)

# save as parquet
aqi.to_parquet('daily_aqi_by_county_2020_2025.parquet', index=False, engine='fastparquet')

print(aqi.shape)
print(aqi.head())
```

## Table 2: Daily NO2 by County from 2020 to 2025

```python
# read all years
no220 = pd.read_csv('daily_42602_2020.csv')
no221 = pd.read_csv('daily_42602_2021.csv')
no222 = pd.read_csv('daily_42602_2022.csv')
no223 = pd.read_csv('daily_42602_2023.csv')
no224 = pd.read_csv('daily_42602_2024.csv')
no225 = pd.read_csv('daily_42602_2025.csv')

# merge all years
no2 = pd.concat([no220, no221, no222, no223, no224, no225], ignore_index=True)

# save as parquet
no2.to_parquet('daily_no2_by_county_2020_2025.parquet', index=False, engine='fastparquet')

print(no2.shape)
print(no2.head())
```

## Table 3: Daily Ozone by County from 2020 to 2025

```python
# read all years
ozone20 = pd.read_csv('daily_44201_2020.csv')
ozone21 = pd.read_csv('daily_44201_2021.csv')
ozone22 = pd.read_csv('daily_44201_2022.csv')
ozone23 = pd.read_csv('daily_44201_2023.csv')
ozone24 = pd.read_csv('daily_44201_2024.csv')
ozone25 = pd.read_csv('daily_44201_2025.csv')

# merge all years
ozone = pd.concat([ozone20, ozone21, ozone22, ozone23, ozone24, ozone25], ignore_index=True)

# save as parquet
ozone.to_parquet('daily_ozone_by_county_2020_2025.parquet', index=False, engine='fastparquet')

print(ozone.shape)
print(ozone.head())
```

## Table 4: Onroad Emissions by Region and County from 2020

```python
# read all regions
reg1 = pd.read_csv('onroad_4.csv')
reg2 = pd.read_csv('onroad_5.csv')
reg3 = pd.read_csv('onroad_67.csv')
reg4 = pd.read_csv('onroad_8910.csv')
reg5 = pd.read_csv('onroad123.csv')

# merge all regions
onroad = pd.concat([reg1, reg2, reg3, reg4, reg5], ignore_index=True)

# save as parquet
onroad.to_parquet('onroad_emissions_by_county_2020.parquet', index=False, engine='fastparquet')

print(onroad.shape)
print(onroad.head())
```
